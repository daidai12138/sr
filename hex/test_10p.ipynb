```python
import os
import re

root_dir = r"E:\A_ligao_model\HEX\experiment\checkpoints"   # 改成你的根目录

results = {}
score=[]
for current_root, dirs, files in os.walk(root_dir):
    folder_name = os.path.basename(current_root)

    # 只处理名字以 aaa 开头的文件夹
    if not folder_name.startswith("20260307_233529"):
        continue

    nums = []

    for fname in files:
        if fname.startswith("checkpoint_best"):
            # 去掉后缀
            stem = os.path.splitext(fname)[0]

            # 取最后一个下划线后的内容
            last_part = stem.split("_")[-1]

            # 如果最后一部分是纯数字
            if last_part.isdigit():
                nums.append(int(last_part))
            else:
                # 更稳妥一点：从最后一部分里提取数字
                m = re.search(r"(\d+(?:\.\d+)?)$", last_part)
                if m:
                    nums.append(float(m.group(1)))
                    score.append(float(m.group(1)))
    results[current_root] = nums

for folder, nums in results.items():
    print(folder, nums)
```

```python
import os
import re

root_dir = r"E:\A_ligao_model\HEX\experiment\checkpoints"
import os
import re
import shutil


threshold = 0.3

for sub_name in os.listdir(root_dir):
    sub_path = os.path.join(root_dir, sub_name)

    # 只处理一级子文件夹
    if not os.path.isdir(sub_path):
        continue

    # 找 .pth 文件
    pth_files = [f for f in os.listdir(sub_path) if f.endswith(".pth")]
    if len(pth_files) == 0:
        print(f"跳过（没有 .pth 文件）: {sub_path}")
        continue

    # 默认每个子文件夹只有一个 .pth
    pth_file = pth_files[0]
    stem = os.path.splitext(pth_file)[0]

    # 取最后一个下划线后的部分
    last_part = stem.split("_")[-1]

    score = None

    # 匹配 r0.3064
    m_pos = re.fullmatch(r"r([0-9]*\.?[0-9]+)", last_part)
    if m_pos:
        score = float(m_pos.group(1))

    # 匹配 rneg0.0686
    m_neg = re.fullmatch(r"rneg([0-9]*\.?[0-9]+)", last_part)
    if m_neg:
        score = -float(m_neg.group(1))

    if score is None:
        print(f"跳过（文件名不符合规则）: {pth_file}")
        continue

    print(f"{sub_path} -> score = {score}")

    if score < threshold:
        shutil.rmtree(sub_path)
        print(f"已删除文件夹: {sub_path}")
```
