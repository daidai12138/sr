```python
import anndata as ad
sample_list = ["N1","N2", "N3", "N4", "N5", "P1", "P2","P3", "P4", "P5"]
sample = 'P1'
adatas=[]
for sample in sample_list:
    adata_data = F'./sample_data/PRO/{sample}/{sample}_pro.h5ad'
    adata = ad.read_h5ad(adata_data)
    adatas.append(adata)
    
adata_merged = ad.concat(
    adatas,
    axis=0,                 # 0=obs(细胞)方向拼接
    join="outer",           # outer=取基因并集；inner=取交集
    label="batch",          # 在 obs 里新增一列 batch
    #keys=[f"s{i}" for i in range(len(adatas))],  # 每个对象的批次名
    index_unique="-",       # 防止 obs_names 重复：如 cell1 -> s0-cell1
    merge="same",           # uns/var/obsm等冲突时的策略（稳妥）
)

adata_merged.obs['img_path'] = (adata_merged.obs["sample_index"].astype(str)+'/'+adata_merged.obs["sample_index"].astype(str)+'_he_'+adata_merged.obs["patch_i"].astype(str)+'.jpg')

adata_merged.write_h5ad("./sample_data/PRO/adata_all.h5ad")
```
